```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
# 1. Create two rank-2 tensors of ones and zeros with 3 rows and columns each, and carry out the arithmetic operations 
#    of addtion, subtraction and multiplication.
```

```python
x = tf.ones([3,3])
x.numpy()
```

```python
y = tf.zeros([3,3])
y.numpy()
```

```python
print(tf.add(x,y))
```

```python
print(tf.subtract(x,y))
```

```python
print(tf.subtract(y,x))
```

```python
print(tf.multiply(x,y))
```

```python
# 2. Suppose you created a rank-3 tf array as follows:
#    array([[[ 6,  2,  4],
#            [12,  8, 10]],

#           [[ 5,  1,  3],
#            [11,  7,  9]],

#          [[21, 32, 54],
#           [13, 49, 36]]], dtype=int32)

#    How to access the following from the above tensor? 
#    array([[[ 2,  4],
#            [ 8, 10]],

#           [[ 1,  3],
#            [ 7,  9]]], dtype=int32)
```

```python
tv = tf.Variable([[[6,2,4],[12,8,10]],[[5,1,3,],[11,7,9]],[[21,32,54],[13,49,36]]])
tv.numpy()
```

```python
tv[0:2,:,1:3].numpy()
```
